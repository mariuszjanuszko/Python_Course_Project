```python
# Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```


ðŸŸ© Drop Missing Values (2.3.1) - Problem
Pandas Data Management
Problem Statement:

Create a copy of the DataFrame and drop rows where the salary_year_avg column has missing values. Display the first 5 rows of the cleaned DataFrame.
Hint:

Use the copy() method to create a copy of the DataFrame and the dropna() method to drop rows with missing values.



```python
df_clean = df.copy()

df_clean = df_clean.dropna(subset='salary_year_avg')
df_clean['salary_year_avg']
```


ðŸŸ© Random Sample of DataFrame (2.3.2) - Problem


ðŸŸ¨ Fill Missing Salaries (2.3.3) - Problem
Pandas Data Management
Problem Statement:

Create a copy of the original DataFrame and fill in missing values in the salary_year_avg column with the median salary. Display the first 5 rows of the altered DataFrame.
Hint:

Use the copy() method to create a copy of the DataFrame and the fillna() method to fill in missing values.


```python
df_filled = df.copy()

df_filled.fillna({'salary_year_avg': df['salary_year_avg'].median()}, inplace=True)
df_filled.head()
```
