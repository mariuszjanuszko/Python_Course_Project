```python
# Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

üü© Concat Job Postings (2.8.1) - Problem
Pandas Concat DataFrames
Problem Statement:

Concatenate two DataFrames to analyze job postings for Data Scientists and Data Analysts:

    Create one DataFrame with job postings for Data Scientists named data_scientist_df and another for Data Analysts named data_scientist_df.
    Concatenate them and display the combined DataFrame called combined_df.

Hint:

Use the pd.concat() function to concatenate the DataFrames.

```python
data_scientist_df = df[df['job_title_short']=='Data Scientist']
data_analyst_df = df[df['job_title_short']=='Data Analyst']
combined_df = pd.concat([data_scientist_df,data_analyst_df], ignore_index=False)
combined_df
```
